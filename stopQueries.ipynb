```python
#11 QUERIES CREATED STARTING FROM OUR COMPETENCY QUESTIONS: 


#which are the topics of the songs?
q1="""
PREFIX rdf: <http://www.w3.org/1999/02/22-rdf-syntax-ns#>
PREFIX stop: <http://www.semanticweb.org/martinapensalfini/ontologies/2022/9/stop#>
SELECT ?topic_name 
WHERE {
    ?song rdf:type stop:Song;
          stop:is_about ?topic.
          ?topic stop:have_name ?topic_name.}"""

#which are the topics of the poems?

q2= """PREFIX rdf: <http://www.w3.org/1999/02/22-rdf-syntax-ns#>
PREFIX stop: <http://www.semanticweb.org/martinapensalfini/ontologies/2022/9/stop#>
SELECT ?topic_name
WHERE {
    ?poem rdf:type stop:Poem;
         stop:is_about ?topic.
         ?topic stop:have_name ?topic_name.
}
"""
#which is the most frequent topic in general?
q3="""PREFIX rdf: <http://www.w3.org/1999/02/22-rdf-syntax-ns#>
PREFIX stop: <http://www.semanticweb.org/martinapensalfini/ontologies/2022/9/stop#>

SELECT ?topic_name (COUNT(?unit) as ?number_of_units)
WHERE {
   ?topic stop:appears_in ?unit.
   ?topic stop:have_name ?topic_name.
   
}
GROUP BY ?topic
ORDER BY desc(?number_of_units)
LIMIT 1

"""
#which is the most frequent topic in songs?
q4="""PREFIX rdf: <http://www.w3.org/1999/02/22-rdf-syntax-ns#>
PREFIX stop: <http://www.semanticweb.org/martinapensalfini/ontologies/2022/9/stop#>

SELECT ?topic_name (COUNT(?unit) as ?number_of_songs)
WHERE {
   ?topic stop:appears_in ?unit.
   ?unit rdf:type stop:Song.
   ?topic stop:have_name ?topic_name.
}
GROUP BY ?topic
ORDER BY desc(?number_of_songs)
LIMIT 1

"""
#which is the most frequent topic in poems?
q5="""PREFIX rdf: <http://www.w3.org/1999/02/22-rdf-syntax-ns#>
PREFIX stop: <http://www.semanticweb.org/martinapensalfini/ontologies/2022/9/stop#>

SELECT ?topic_name (COUNT(?unit) as ?number_of_poems)
WHERE {
   ?topic stop:appears_in ?unit.
   ?unit rdf:type stop:Poem.
   ?topic stop:have_name ?topic_name.
}
GROUP BY ?topic
ORDER BY desc(?number_of_poems)
LIMIT 1

"""

#which are the topics treated in the song "Un chimico"? 
q6="""
 PREFIX rdf: <http://www.w3.org/1999/02/22-rdf-syntax-ns#>
PREFIX stop: <http://www.semanticweb.org/martinapensalfini/ontologies/2022/9/stop#>

SELECT ?topic_name

WHERE {
    ?unit rdf:type stop:Song;
        stop:have_name "Un chimico"^^rdfs:Literal;
         stop:is_about ?topic.
         ?topic stop:have_name ?topic_name.
    
}
"""
#q7which are the topics treated in the poem "Trainor the druggist"?
q7="""PREFIX rdf: <http://www.w3.org/1999/02/22-rdf-syntax-ns#>
PREFIX stop: <http://www.semanticweb.org/martinapensalfini/ontologies/2022/9/stop#>
SELECT ?topic_name

WHERE {
    ?unit rdf:type stop:Poem;
        stop:have_name "Trainor, the druggist"^^rdfs:Literal;
         stop:is_about ?topic.
         ?topic stop:have_name ?topic_name.
    
}
"""

#which songs/poems are associated with the topic "Sky"?

q8="""
 PREFIX rdf: <http://www.w3.org/1999/02/22-rdf-syntax-ns#>
PREFIX stop: <http://www.semanticweb.org/martinapensalfini/ontologies/2022/9/stop#>

SELECT ?unit_name
WHERE {
  VALUES ?type { stop:Poem stop:Song }
    ?unit rdf:type ?type;
          stop:have_name ?unit_name;
            stop:is_about stop:Sky.
}
"""
#in which songs/poems can we find the topic "Love"?
q9="""PREFIX rdf: <http://www.w3.org/1999/02/22-rdf-syntax-ns#>
PREFIX stop: <http://www.semanticweb.org/martinapensalfini/ontologies/2022/9/stop#>

SELECT ?unit_name
WHERE {
    VALUES ?type { stop:Poem stop:Song }
    ?unit rdf:type ?type;
          stop:have_name ?unit_name;
            stop:is_about stop:Love.}
"""


#which characters of The hill are not mentioned in La collina?
q10="""PREFIX rdf: <http://www.w3.org/1999/02/22-rdf-syntax-ns#>
PREFIX stop: <http://www.semanticweb.org/martinapensalfini/ontologies/2022/9/stop#>
SELECT  ?p_name
WHERE {        
  ?song stop:have_name "La collina"^^rdfs:Literal;
        stop:related_to ?poem.
  ?poem stop:has_character ?p_character.
  ?p_character stop:have_name ?p_name.
  FILTER (NOT EXISTS {?song stop:has_character ?p_character})
  }
"""

#which is the name of the the original poem by Masters that inspired the song "Un giudice"?
q11="""
 PREFIX rdf: <http://www.w3.org/1999/02/22-rdf-syntax-ns#>
PREFIX stop: <http://www.semanticweb.org/martinapensalfini/ontologies/2022/9/stop#>

SELECT ?poem_name
WHERE {
?unit stop:have_name "Un giudice"^^rdfs:Literal;
      rdf:type stop:Song;
      stop:related_to ?poem.
?poem stop:have_name ?poem_name.}
"""

queries = [q1,q2,q3,q4,q5,q6,q7,q8,q9,q10,q11]

for idx, q in enumerate(queries):
    print("Results for query " + str(idx + 1) + ":")
    results = spoonR.query(q)
    for result in results:
        print(result) 
```
